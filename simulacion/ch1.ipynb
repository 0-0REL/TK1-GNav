Simulate a vehicle with 1.5 m length, steering 45° and 2.778 m/s

```python
alpha = math.radians(45)
L = 1.5
V = 2.778
T = 0.001 # sampling interval (seconds)
SIM_T = 3.33
KT = int(SIM_T / T)
x = np.zeros(KT)
y = np.zeros(KT)
psi = np.zeros(KT)
for k in range(KT - 1):
    x[k + 1] = x[k] - T * V * math.sin(psi[k])
    y[k + 1] = y[k] + T * V * math.cos(psi[k])
    psi[k + 1] = psi[k] + T * V / L * math.tan(alpha)
plt.plot(x, y)
plt.title("trajectory")
plt.grid()
plt.show()
```

Simulation vehicle has 1 m wide and 2 m length

inputs: speed and steering (+-45°)

Straight line, steering 45° constant, steering -45° constant

```python
L = 2
T = 0.1 # sampling interval
SIM_T = 100 # simulation time
KT = int(SIM_T / T)
x = np.zeros(KT)
y = np.zeros(KT)
psi = np.zeros(KT)
alpha = np.zeros(KT)
v = np.zeros(KT)

for k in range(KT - 1):
    if k < KT//3:
        u1 = 0 # rad/s
        u2 = 0.01 # m/s²
    elif k >= KT//3 and k < 2*KT//3:
        u1 = math.radians(10)
        u2 = 0
    else:
        u1 = math.radians(-10)
        u2  = 0
    x[k + 1] = x[k] - T * v[k] * math.sin(psi[k])
    y[k + 1] = y[k] + T * v[k] * math.cos(psi[k])
    psi[k + 1] = psi[k] + T * v[k] / L * math.tan(alpha[k])
    alpha[k + 1] = alpha[k] + T * u1
    v[k + 1] = v[k] + T * u2
    if alpha[k + 1] > math.radians(45):
        alpha[k + 1] = math.radians(45)
    elif alpha[k + 1] < math.radians(-45):
        alpha[k + 1] = math.radians(-45)
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x)
axs[0,0].set_title('x vs t')
axs[1,0].plot(y)
axs[1,0].set_title('y vs t')
axs[0,1].plot(psi)
axs[0,1].set_title('psi vs t')
axs[1,1].plot(x,y)
axs[1,1].set_title('x vs y')
```
